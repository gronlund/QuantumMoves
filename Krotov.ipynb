# Krotov Bring Home Water
This small notebook shows how to run the basic Krotov algorithm on the Bring Home Water problem.

```python
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import krotov
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

## Parameters to play with
Below i have listed the parameters to play with
- T is duration to consider
- N is number of steps to do
- hamiltonian_grid_size: the size of the grid used to discretize the hamiltonian operator - large means more accurate but slower algorithm - but we set it small here because the algorithm becomes very slow
- A Max amplitude of controllable tweezer  (160)
- B amplitude of fixed tweezer (130)

We have fixed some standard value below and run the stochastic ascent algorithm that prints out what it is doing.
We have on purpose used smaller hamiltonian_grid_size compared to the experiments in the note because it speeds everything up and works well enough.
In the follwing cell we run the stochastic ascent algorithm with these parameters. 

```python
T = 0.2
N = 80
hamiltonian_grid_size = 32
A = 160
B = 130
params = {'T': T, 'N': N, 'hamiltonian_grid_size': hamiltonian_grid_size, 
          'Alpha': A, 'Beta': B}

rounds = 1000
lr_pos = 0.01
lr_amp = 0.1

```

## Completely Uniform Random Initalization 
We run 1000 steps of the krotov algorithm - which is pretty fast actually.
**Every give  different results as we have random initialization**. 
It should be clear that we get much better fidelity at duration 0.2 than the KASS algorithm (also based on Krotov)!

```python
start_amplitudes = np.random.uniform(size=(N,), low=0.0, high=160.0)  
start_positions = np.random.uniform(size=(N,), low=-1.0, high=1.0)  
tf.reset_default_graph()
best_fid, best_pos, best_amp, scores = krotov.krotov(params, start_positions=start_positions, start_amplitudes=start_amplitudes, 
                                                     lr_pos=lr_pos, lr_amp=lr_amp, rounds=rounds)
```

## Wauw not bad for duration 0.2 when compared to KASS! 
### Lets plot the protocol and see what it looks like


```python
def plot_protocol(position_dict, amplitude_dict, params):
    """ Plot protocol that uses time t """
    N = params['N']
    t = params['T']
    positions = [position_dict[i] for i in range(1, N+1)]
    amplitudes = [amplitude_dict[i] for i in range(1, N+1)]
    
    fig, axes = plt.subplots(1, 2, figsize=(32, 16), sharex=True)
    delta_t = t/len(positions)
    xr = delta_t + np.array(list(range(len(positions)))) * delta_t
    axes[0].step(xr,  positions, '-', color='green', linewidth=2, markerfacecolor="None", markersize=6)
    axes[0].set_ylim([-1, 1])
    axes[0].set_xlabel('Time', fontsize=16)
    axes[0].set_ylabel('Tweezer Positions', fontsize=16)
    axes[0].set_title('Position Plot', fontsize=16)
    axes[1].step(xr,  amplitudes, '-', color='green', linewidth=2, markerfacecolor="None", markersize=6)
    axes[1].set_ylim([0, 165])
    axes[1].set_xlabel('Time', fontsize=16)
    axes[1].set_ylabel('Tweezer Amplitude', fontsize=16)
    axes[1].set_title('Amplitude Plot', fontsize=16)

    
plot_protocol(best_pos, best_amp, params)
plt.show()
```

## Numerical Accuracy  
We used hgrid=32 which means the numerical accuracy in the fidelity computation could be better.
So lets make it better. (We reduce the printing below).


```python
params_64 = {'T': T, 'N': N, 'hamiltonian_grid_size': 64, 'Alpha': A, 'Beta': B}
params_128 = {'T': T, 'N': N, 'hamiltonian_grid_size': 128, 'Alpha': A, 'Beta': B}
params_256 = {'T': T, 'N': N, 'hamiltonian_grid_size': 256, 'Alpha': A, 'Beta': B}
params_512 = {'T': T, 'N': N, 'hamiltonian_grid_size': 512, 'Alpha': A, 'Beta': B}

lr_pos = 0.01
lr_amp = 0.1

tf.reset_default_graph()
best_fid_64, best_pos_64, best_amp_64, _ = krotov.krotov(params_64, start_positions=best_pos, start_amplitudes=best_amp, 
                                                         lr_pos=lr_pos, lr_amp=lr_amp, rounds=500, verbose=0)
print('Fidelity with hgrid 64', best_fid_64)
tf.reset_default_graph()
best_fid_128, best_pos_128, best_amp_128, _ = krotov.krotov(params_128, start_positions=best_pos_64, start_amplitudes=best_amp_64, 
                                                         lr_pos=lr_pos, lr_amp=lr_amp, rounds=200, verbose=0)
print('Fidelity with hgrid 128', best_fid_128)
tf.reset_default_graph()
best_fid_256, best_pos_256, best_amp_256, _ = krotov.krotov(params_256, start_positions=best_pos_128, start_amplitudes=best_amp_128, 
                                                         lr_pos=lr_pos, lr_amp=lr_amp, rounds=50, verbose=0)
print('Fidelity with hgrid 256', best_fid_256)
tf.reset_default_graph()
best_fid_512, best_pos_512, best_amp_512, _ = krotov.krotov(params_512, start_positions=best_pos_256, start_amplitudes=best_amp_256, 
                                                         lr_pos=lr_pos, lr_amp=lr_amp, rounds=5, verbose=0)
print('Fidelity with hgrid 512', best_fid_512)
```
